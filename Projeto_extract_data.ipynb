```python
def extract_urls(url):
        
    from selenium import webdriver
    from selenium.webdriver.common.keys import Keys
    from tqdm.auto import tqdm
    chrome_path = 'C:/Users/youngbae/Downloads/chromedriver_win32/chromedriver'
    driver = webdriver.Chrome(executable_path=chrome_path)
    driver.get(url)
    button=driver.find_element_by_xpath('/html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[2]/section/ais-infinite-hits/div/div[1]/div/div[1]')
    button.click()
    html = driver.find_element_by_tag_name('html')
    for i in tqdm(range(1,800)):
        html.send_keys(Keys.END)
    elem=driver.find_elements_by_tag_name('a')
    elem=[i.get_attribute('href') for i in elem if i.get_attribute('href')!=None]
    urls=[i for i in elem if i.startswith('https://startupbase.com.br/c/startup/')]
    driver.close()
    return urls

```

```python
def transform(data):
    import pandas as pd
    import numpy as np
    
    df=pd.DataFrame(data)
    df.columns=('Name','Estado/Cidade','Segmento','Público Alvo','Modelo de Receita','Ano de Fundação','Tamanho','Mais Informação','Momento')
    df=df.dropna(how='any')
    cidade_estado=[i.split(' - ') for i in list(df['Estado/Cidade'])]
    df_estado_cidade=pd.DataFrame(cidade_estado)
    df_estado_cidade.columns=['Cidade','Estado']
    df_final=pd.concat([df, df_estado_cidade], axis=1)
    df_final=df_final[['Name','Estado','Cidade','Segmento','Público Alvo','Modelo de Receita','Ano de Fundação','Tamanho','Mais Informação','Momento']]
    return df_final
```
